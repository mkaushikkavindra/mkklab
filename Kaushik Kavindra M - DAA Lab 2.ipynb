```
import time
import matplotlib.pyplot as plt
import random

def binary_search_recursive(arr, target, low, high):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search_recursive(arr, target, low, mid - 1)
    else:
        return binary_search_recursive(arr, target, mid + 1, high)

def run_experiment():
    sizes = [1000 * i for i in range(1, 21)]
    times = []

    for n in sizes:
        arr = sorted(random.sample(range(0, n * 10), n))
        target = random.choice(arr)

        start_time = time.perf_counter()
        binary_search_recursive(arr, target, 0, len(arr) - 1)
        end_time = time.perf_counter()

        elapsed_time = end_time - start_time
        times.append(elapsed_time)

    plt.figure(figsize=(10, 5))
    plt.plot(sizes, times, marker='o', linestyle='-', color='blue')
    plt.title('Recursive Binary Search: Time vs List Size')
    plt.xlabel('List Size (n)')
    plt.ylabel('Time Taken (seconds)')
    plt.grid(True)
    plt.show()

run_experiment()
```
