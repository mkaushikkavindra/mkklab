**Lab Question 1: Performance Analysis of Linear Search on Best, Average, and Worst Cases**

**Problem Statement:**
Write a program to implement the linear search algorithm. Analyze and compare the time taken for:
* Best case (element at the first position),
* Average case (element in the middle),
* Worst case (element not present).

**Instructions:**
* Generate a list of n random integers (e.g., n = 10000).
* Measure the time taken for searching the key in each of the three cases.
* Plot a graph comparing the time taken vs. the case type.

**Expected Analysis Output:**
* Explain why the best case is O(1), average case is O(n/2), and worst case is O(n).
* Include actual execution times in milliseconds.
* Provide observations on the impact of input size n.

```
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def measure_time(n):
    arr = [random.randint(1, 10000) for _ in range(n)]

    best_case_key = arr[0]
    start_time = time.time()
    linear_search(arr, best_case_key)
    best_case_time = (time.time() - start_time) * 1000

    average_case_key = arr[n // 2]
    start_time = time.time()
    linear_search(arr, average_case_key)
    average_case_time = (time.time() - start_time) * 1000

    worst_case_key = -1
    start_time = time.time()
    linear_search(arr, worst_case_key)
    worst_case_time = (time.time() - start_time) * 1000

    return best_case_time, average_case_time, worst_case_time

def analyze_performance():
    n_values = [100, 1000, 5000, 10000, 20000]
    best_case_times = []
    average_case_times = []
    worst_case_times = []

    for n in n_values:
        best, avg, worst = measure_time(n)
        best_case_times.append(best)
        average_case_times.append(avg)
        worst_case_times.append(worst)

    plt.plot(n_values, best_case_times, label="Best Case", marker='o')
    plt.plot(n_values, average_case_times, label="Average Case", marker='o')
    plt.plot(n_values, worst_case_times, label="Worst Case", marker='o')

    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (ms)')
    plt.title('Linear Search: Time Complexity Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

analyze_performance()
```

**Lab Question 2: Count the Number of Comparisons in Linear Search**

**Problem Statement:**
Implement linear search and count the number of comparisons made during the search process.
For various input sizes (e.g., n = 100, 500, 1000, 5000, 10000),determine:

* Total comparisons in best, average, and worst cases.
* Plot the results to show growth of comparisons with input size.

**Instructions:**
* Run your program multiple times with varying input sizes.
* For each size, simulate best, average, and worst-case conditions.
* Count and display the number of comparisons made.

**Expected Analysis Output:**
* Tabulate input size vs. number of comparisons.
* Explain how the number of comparisons grows linearly.
* Compare empirical results with theoretical expectations (O(1), O(n/2), O(n).

```
import random
import matplotlib.pyplot as plt

def linear_search_count(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return comparisons
    return comparisons

def measure_comparisons(n):
    arr = [random.randint(1, 10000) for _ in range(n)]

    best_case_key = arr[0]
    best_case_comparisons = linear_search_count(arr, best_case_key)

    average_case_key = arr[n // 2]
    average_case_comparisons = linear_search_count(arr, average_case_key)

    worst_case_key = -1
    worst_case_comparisons = linear_search_count(arr, worst_case_key)

    return best_case_comparisons, average_case_comparisons, worst_case_comparisons

def analyze_performance():
    n_values = [100, 500, 1000, 5000, 10000]
    best_case_comparisons = []
    average_case_comparisons = []
    worst_case_comparisons = []

    for n in n_values:
        best, avg, worst = measure_comparisons(n)
        best_case_comparisons.append(best)
        average_case_comparisons.append(avg)
        worst_case_comparisons.append(worst)

    plt.plot(n_values, best_case_comparisons, label="Best Case", marker='o')
    plt.plot(n_values, average_case_comparisons, label="Average Case", marker='o')
    plt.plot(n_values, worst_case_comparisons, label="Worst Case", marker='o')

    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Linear Search: Number of Comparisons vs. Input Size')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"{'n':<10}{'Best Case Comparisons':<20}{'Average Case Comparisons':<25}{'Worst Case Comparisons':<25}")
    for i in range(len(n_values)):
        print(f"{n_values[i]:<10}{best_case_comparisons[i]:<20}  {average_case_comparisons[i]:<25}  {worst_case_comparisons[i]:<25}")

analyze_performance()
```
